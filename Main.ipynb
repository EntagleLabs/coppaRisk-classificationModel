Melihat fitur mana yg paling berpengaruh setelah di One Hot Encoding menggunakan library SHAP

```python
import pandas as pd
import numpy as np
from lightgbm import LGBMClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
import shap

# === Load Data ===
train_df = pd.read_csv("dataset/cleaned_data(1).csv")
test_df = pd.read_csv("dataset/edit_test.csv")
target = 'coppaRisk'
X_train = train_df.drop(columns=[target])
y_train = train_df[target]
test_ids = test_df["ID"]
X_test = test_df.drop(columns=["ID"])

# === Preprocessing Pipeline ===
numeric_cols = X_train.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = X_train.select_dtypes(include=["object", "bool"]).columns.tolist()

numeric_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocessor = ColumnTransformer([
    ("num", numeric_transformer, numeric_cols),
    ("cat", categorical_transformer, categorical_cols)
])

# === Transform Data ===
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)
X_train_dense = X_train_transformed if hasattr(X_train_transformed, "toarray") else X_train_transformed
X_test_dense = X_test_transformed if hasattr(X_test_transformed, "toarray") else X_test_transformed

# === Feature Names ===
onehot_names = preprocessor.named_transformers_['cat'].named_steps['encoder'].get_feature_names_out(categorical_cols)
all_features = np.concatenate([numeric_cols, onehot_names])

# === SHAP Calculation ===
lgbm_model = LGBMClassifier(random_state=42)
lgbm_model.fit(X_train_dense, y_train)
explainer = shap.TreeExplainer(lgbm_model)
shap_values = explainer.shap_values(X_train_dense)
shap_to_plot = shap_values[1] if isinstance(shap_values, list) else shap_values
shap_mean = np.abs(shap_to_plot).mean(axis=0)

# === Top Features ===
top50_idx = np.argsort(shap_mean)[-50:][::-1]
top30_idx = np.argsort(shap_mean)[-30:][::-1]
top50_features = [all_features[i] for i in top50_idx]
top30_features = [all_features[i] for i in top30_idx]

# === Output DataFrames ===
X_train_top50 = pd.DataFrame(X_train_dense, columns=all_features)[top50_features]
X_test_top50 = pd.DataFrame(X_test_dense, columns=all_features)[top50_features]
X_train_top30 = pd.DataFrame(X_train_dense, columns=all_features)[top30_features]
X_test_top30 = pd.DataFrame(X_test_dense, columns=all_features)[top30_features]

# === Save (opsional) ===
X_train_top50.to_csv("X_train_top50.csv", index=False)
X_train_top30.to_csv("X_train_top30.csv", index=False)
X_test_top50.to_csv("X_test_top50.csv", index=False)
X_test_top30.to_csv("X_test_top30.csv", index=False)
```

```python
# menampilkan Top 30 SHAP value tertinggi beserta nama fiturnya
shap_summary = pd.Series(shap_mean, index=all_features)
top_shap_values = shap_summary.sort_values(ascending=False).head(30)

top_shap_values.plot(kind='barh', figsize=(10, 8))
plt.title("Top 30 SHAP Feature Importances")
plt.xlabel("Mean |SHAP Value|")
plt.gca().invert_yaxis()  
plt.tight_layout()
plt.show()
```

Model Menggunakan algoritma LightGBM 

Menggunakan Top 30 fitur yang sudah di tuning menggunakan hyperparameter lalu di assemble dengan model yg menggunakan Top 50 fitur 

```python
# model Top 30 tuned + top 50 SHap
import pandas as pd
import numpy as np
from lightgbm import LGBMClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import StratifiedKFold, cross_val_score, RandomizedSearchCV
from sklearn.metrics import roc_auc_score
import shap

# === Load Data ===
train_df = pd.read_csv("dataset/cleaned_data(1).csv")
test_df = pd.read_csv("dataset/edit_test.csv")
target = 'coppaRisk'
X_train = train_df.drop(columns=[target])
y_train = train_df[target]
test_ids = test_df["ID"]
X_test = test_df.drop(columns=["ID"])

# === Preprocessing Pipeline ===
numeric_cols = X_train.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = X_train.select_dtypes(include=["object", "bool"]).columns.tolist()

numeric_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocessor = ColumnTransformer([
    ("num", numeric_transformer, numeric_cols),
    ("cat", categorical_transformer, categorical_cols)
])

# === Transform Data ===
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)
X_train_dense = X_train_transformed if isinstance(X_train_transformed, np.ndarray) else X_train_transformed.toarray()
X_test_dense = X_test_transformed if isinstance(X_test_transformed, np.ndarray) else X_test_transformed.toarray()

# === Feature Names ===
onehot_names = preprocessor.named_transformers_['cat'].named_steps['encoder'].get_feature_names_out(categorical_cols)
all_features = np.concatenate([numeric_cols, onehot_names])

# === SHAP Calculation ===
lgbm_model = LGBMClassifier(random_state=42)
lgbm_model.fit(X_train_dense, y_train)
explainer = shap.TreeExplainer(lgbm_model)
shap_values = explainer.shap_values(X_train_dense)
shap_to_plot = shap_values[1] if isinstance(shap_values, list) else shap_values
shap_mean = np.abs(shap_to_plot).mean(axis=0)

# === Top-50 SHAP Model ===
top50_idx = np.argsort(shap_mean)[-50:][::-1]
top50_features = [all_features[i] for i in top50_idx]
X_train_top50 = pd.DataFrame(X_train_dense, columns=all_features)[top50_features]
X_test_top50 = pd.DataFrame(X_test_dense, columns=all_features)[top50_features]

model_top50 = LGBMClassifier(random_state=42)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_auc_top50 = cross_val_score(model_top50, X_train_top50, y_train, cv=cv, scoring="roc_auc")
model_top50.fit(X_train_top50, y_train)
y_pred_top50 = model_top50.predict_proba(X_test_top50)[:, 1]

# === Top-30 SHAP Model ===
top30_idx = np.argsort(shap_mean)[-30:][::-1]
top30_features = [all_features[i] for i in top30_idx]
X_train_top30 = pd.DataFrame(X_train_dense, columns=all_features)[top30_features]
X_test_top30 = pd.DataFrame(X_test_dense, columns=all_features)[top30_features]

# === Hyperparameter Tuning (Top-30) ===
param_grid = {
    'num_leaves': [31, 63, 127],
    'max_depth': [-1, 7, 10, 15],
    'learning_rate': [0.005, 0.01, 0.05, 0.1],
    'n_estimators': [100, 200, 500],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
}
random_search = RandomizedSearchCV(
    estimator=LGBMClassifier(random_state=42),
    param_distributions=param_grid,
    n_iter=30,
    scoring='roc_auc',
    cv=cv,
    n_jobs=-1,
    verbose=1,
    random_state=42
)
random_search.fit(X_train_top30, y_train)
best_model_top30 = random_search.best_estimator_
cv_auc_top30_tuned = cross_val_score(best_model_top30, X_train_top30, y_train, cv=cv, scoring="roc_auc")
best_model_top30.fit(X_train_top30, y_train)
y_pred_top30 = best_model_top30.predict_proba(X_test_top30)[:, 1]

# === Ensemble Prediction (Top-50 + Top-30 Tuned) ===
y_pred_ensemble = 0.7 * y_pred_top50 + 0.3 * y_pred_top30

# === Save Submissions ===
submission_top50 = pd.DataFrame({"ID": test_ids, "coppaRisk": y_pred_top50})
submission_top30 = pd.DataFrame({"ID": test_ids, "coppaRisk": y_pred_top30})
submission_ensemble = pd.DataFrame({"ID": test_ids, "coppaRisk": y_pred_ensemble})

submission_top50.to_csv("submission_top50.csv", index=False)
submission_top30.to_csv("submission_top30_tuned.csv", index=False)
submission_ensemble.to_csv("submission_ensemblev2.csv", index=False)

# === Print AUC Results ===
print("\n Mean AUC CV (Top-50):", np.mean(cv_auc_top50))
print(" AUC per Fold:", cv_auc_top50)

print("\n Mean AUC CV (Top-30 TUNED):", np.mean(cv_auc_top30_tuned))
print(" AUC per Fold:", cv_auc_top30_tuned)

print("\n Semua file submission telah disiapkan dan tuning model berhasil dilakukan.")
```
